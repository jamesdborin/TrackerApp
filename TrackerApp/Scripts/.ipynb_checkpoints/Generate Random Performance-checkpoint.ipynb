```python
"""
we need to generate a file with the form:

Date1 
WorkoutID1 
ExerciseName1 Performance1
ExerciseName2 Performance2
...

Date2
WorkoutID2
ExerciseName1 Performance1
ExerciseName2 Performance2
"""
import numpy as np
from datetime import date, timedelta
from TrackerApp.ExerciseNetwork import ExerciseNetwork
today = date.today()
day = timedelta(days = 1)


ex = ExerciseNetwork()
ex.file_to_graph("../stored_info/first_layer_exercise_chart.txt")
exercises = ex.get_exercise_names()
sets = [3,4,5]
reps = [3,5,8,10,12]
num_exercises = [4,5,6]

"""
To simulate working out over several months:

Choose length of simulation

Decide if you workout

Decide how many exercises you do

Pick random exercises 

For each exercise choose a random number of sets and a random number of reps

Choose weights as a normal distribution around 50 with a range of 5.

Turn into a performance string.

Save as a string
"""
```

```python
string = ''
workout = 1
simulation_length = 500
date = today
for _ in range(simulation_length):
    do_workout = np.random.randint(0,7) > 2
    if do_workout:
        string += (str(date) + " \n")
        num_ex = np.random.choice(num_exercises)

        string+=(str(workout) + "\n")
        s = np.random.choice(sets, num_ex)
        r = np.random.choice(reps, num_ex)
        e = np.random.choice(exercises, num_ex, replace = False)
        
        for s_,r_,e_ in zip(s,r,e):
            string += str(e_) + ":"
            w = 5 * np.round(int(np.random.randn(1)[0])) + 50
            performance = ','.join([f"{r_}@{w}"]*s_)
            string += " " + performance + "\n"
        
        string += "\n"
        date+=day
        workout += 1
    else:
        date += day
        continue
```

```python
%store string >simulate_workout.txt

```
